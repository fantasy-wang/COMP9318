```python
import pandas as pd
import numpy as np
```

```python
info = pd.DataFrame({'Time': ['05/03/2020'], 
                     'Deposit': [400], 
                     'Account balance': [60], 
                     'Withdraw': [0], 
                     'Single_profit': [-340],
                     'Total_profit': [-340]})

info[['Deposit', 'Account balance', 'Withdraw']] = info[['Deposit', 'Account balance', 'Withdraw']].apply(pd.to_numeric)
info
```

```python
def note_and_profit(info, time, deposit, account_balance, withdraw):
    total_1 = info['Deposit'].sum() + deposit
    total_2 = info['Withdraw'].sum() + withdraw
    profit = account_balance + total_2 - total_1
    single = profit - info['Total_profit'][len(info['Total_profit']) - 1]
    info = info.append({'Time': time, 'Deposit': deposit, 
                        'Account balance': account_balance, 
                        'Withdraw': withdraw,
                        'Single_profit': single,
                        'Total_profit': profit}, 
                       ignore_index = True)
    
    return info

```

```python
info = note_and_profit(info, '06/03/2020', 340, 400, 800)
info
```

```python
info = note_and_profit(info, '07/03/2020', 0, 400, 633)
info
```

```python
info = note_and_profit(info, '08/03/2020', 0, 0, 0)
info
```

```python
info = note_and_profit(info, '09/03/2020', 800, 400, 1315)
info
```

```python
info = note_and_profit(info, '09/03/2020', 0, 400, 926)
info
```
